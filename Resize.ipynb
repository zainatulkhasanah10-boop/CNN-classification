```python
from google.colab import files
import cv2

# Upload file gambar
uploaded = files.upload()

# Ambil nama file yang diupload
img_path = list(uploaded.keys())[0]

# Baca gambar
img = cv2.imread(img_path)

# Cek apakah gambar berhasil dibaca
if img is None:
    print("Gambar tidak ditemukan. Periksa path file.")
else:
    print("Gambar berhasil dibaca.")
def resize_with_padding(image, target_size):
    old_size = image.shape[:2]  # Dimensi asli (height, width)
    ratio = float(target_size) / max(old_size)  # Rasio untuk resize
    new_size = tuple([int(x * ratio) for x in old_size])  # Ukuran baru sesuai rasio

    # Resize gambar
    resized_image = cv2.resize(image, (new_size[1], new_size[0]))

    # Hitung padding yang dibutuhkan
    delta_w = target_size - new_size[1]
    delta_h = target_size - new_size[0]

    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)

    # Tambahkan padding (warna padding default: hitam)
    padded_image = cv2.copyMakeBorder(resized_image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])

    return padded_image

# Resize gambar dengan padding
if img is not None:
    target_size = 224  # Ukuran target
    final_image = resize_with_padding(img, target_size)

    # Simpan gambar hasil resize
    cv2.imwrite('resized_image_with_padding.png', final_image)
else:
    print("Gambar tidak tersedia untuk di-resize.")
```

**RESIZE Semua Folder**

```python
import cv2
import os
import glob
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Tentukan path folder input dan output untuk setiap jenis padi
input_folders = {
    'ciherang': '/content/drive/MyDrive/CNN/Padi/Augmentasi/BR_1/Ciherang/',
    'inpari': '/content/drive/MyDrive/CNN/Padi/Augmentasi/BR_1/Inpari/',
    'sintanur': '/content/drive/MyDrive/CNN/Padi/Augmentasi/BR_1/Sintanur/'
}

output_folders = {
    'ciherang': '/content/drive/MyDrive/CNN/Padi/Resize/Setelah/BR_1/Ciherang/',
    'inpari': '/content/drive/MyDrive/CNN/Padi/Resize/Setelah/BR_1/Inpari/',
    'sintanur': '/content/drive/MyDrive/CNN/Padi/Resize/Setelah/BR_1/Sintanur/'
}

# Membuat folder output jika belum ada
for folder in output_folders.values():
    os.makedirs(folder, exist_ok=True)

# Fungsi untuk resize dengan padding
def resize_with_padding(image, target_size=224):
    old_size = image.shape[:2]  # Dimensi asli (height, width)
    ratio = float(target_size) / max(old_size)  # Rasio untuk resize
    new_size = tuple([int(x * ratio) for x in old_size])  # Ukuran baru sesuai rasio

    # Resize gambar
    resized_image = cv2.resize(image, (new_size[1], new_size[0]))

    # Hitung padding yang dibutuhkan
    delta_w = target_size - new_size[1]
    delta_h = target_size - new_size[0]
    top, bottom = delta_h // 2, delta_h - (delta_h // 2)
    left, right = delta_w // 2, delta_w - (delta_w // 2)

    # Tambahkan padding (warna padding default: hitam)
    padded_image = cv2.copyMakeBorder(resized_image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=[0, 0, 0])

    return padded_image

# Proses setiap gambar dalam folder input
for key, input_folder in input_folders.items():
    # Dapatkan folder output yang sesuai
    output_folder = output_folders[key]

    # Ambil semua file gambar dalam folder (format .jpg dan .png)
    image_paths = glob.glob(os.path.join(input_folder, '*.jpg')) + glob.glob(os.path.join(input_folder, '*.png'))

    for img_path in image_paths:
        # Baca gambar
        img = cv2.imread(img_path)

        # Cek apakah gambar berhasil dibaca
        if img is None:
            print(f"Gambar tidak ditemukan: {img_path}. Periksa path file.")
            continue  # Lanjutkan ke gambar berikutnya

        print(f"Gambar berhasil dibaca: {img_path}")

        # Resize gambar dengan padding
        final_image = resize_with_padding(img, target_size=224)

        # Siapkan path untuk menyimpan gambar hasil resize
        output_path = os.path.join(output_folder, os.path.basename(img_path))

        # Simpan gambar hasil resize
        cv2.imwrite(output_path, final_image)
        print(f"Gambar disimpan: {output_path}")
```
