```python
import os
import shutil
import random
from google.colab import drive


# Folder asal data
data_folder = '/content/drive/MyDrive/BagiData'
categories = ['Ciherang', 'Inpari', 'Sintanur']  # Nama kategori
split_ratio = 0.7  # 70% untuk training

# Membuat folder untuk data training dan testing
train_folder = '/content/drive/MyDrive/BagiData/Training'
test_folder = '/content/drive/MyDrive/BagiData/Testing'

os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

for category in categories:
    # Membaca gambar dari setiap kategori
    category_folder = os.path.join(data_folder, category)
    #random
    images = [f for f in os.listdir(category_folder) if f.endswith('.png') or f.endswith('.jpg')]
    random.shuffle(images) # Mengacak daftar gambar
    #Urut
    #images = sorted([f for f in os.listdir(category_folder) if f.endswith('.png') or f.endswith('.jpg')])

    # Menghitung batas untuk pembagian data
    train_size = int(len(images) * split_ratio)

    # Membagi data menjadi training dan testing
    train_images = images[:train_size]
    test_images = images[train_size:]

    # Membuat folder untuk setiap kategori dalam training dan testing
    os.makedirs(os.path.join(train_folder, category), exist_ok=True)
    os.makedirs(os.path.join(test_folder, category), exist_ok=True)

    # Menyimpan gambar ke folder yang sesuai
    for img in train_images:
        shutil.copy(os.path.join(category_folder, img), os.path.join(train_folder, category, img))

    for img in test_images:
        shutil.copy(os.path.join(category_folder, img), os.path.join(test_folder, category, img))

print("Pembagian data selesai!")
```
